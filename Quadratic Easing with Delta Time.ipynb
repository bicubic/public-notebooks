```python
def generate_easing(steps=10):
    end_time = 1
    start = 0.0
    end = 1.0
    time_step = end / (steps)

    points = []
    current = start
    time = 0
    for i in range(steps):
        delta_position = end - current
        time += time_step
        current += delta_position * 3.0 * time_step
        points.append((time, current))
    return points
```

```python
for res in [10,20,100,1000]:
    points = generate_easing(res)
    x, y = zip(*points)
    plt.plot(x,y, label='{0} Points'.format(res))


plt.legend(loc='lower right');
```
